```python
!pip install requests
import requests
res = requests.get("https://www.esportsearnings.com/history/2021/teams")
```

```python
soup = bs(res.content)
```

```python
# prettify is a function that returns a nicely formatted HTML with newlines and indent
# that makes it easier to read
print(soup.prettify())
```

```python
data = []
table = soup.find('table', attrs={'class':'detail_list_table'})
table_body = table.find('tbody')

rows = table_body.find_all('tr')
for row in rows:
    cols = row.find_all('td')
    cols = [ele.text.strip() for ele in cols]
    data.append([ele for ele in cols if ele]) # Get rid of empty values
```

```python
data
```

```python
import pandas as pd 
esports_df = pd.DataFrame(data, columns=cols)
```

```python
esports_df
```

```python
esports_df.columns
```

```python
df= esports_df.rename(columns={"485.":"RankOrder",})
```

```python
df= df.rename(columns={"LANGUDAR":"TeamName",
                               "$0.00":"TournamentPrize",
                               "1 Tournament":"TournamentsPlayed"})
```
